```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.patches as patch

%matplotlib inline
rcParams['figure.figsize'] = [10., 10.]
```

```python
data1 = pandas.read_csv('Downloads/alphap_data.csv')

fig, ax = plt.subplots()

# Alpha Perei
aP1 = data1.query('parallax < 8 & parallax > 4')    
n, bins, patches = ax.hist(aP1['parallax'], bins = 100)
ax.set_xlabel('Parallax [mas]')
ax.set_ylabel('Star Counts')
plt.vlines(5.31,0, 250, color='k',linestyle='--', label = 'Estimated Width')
plt.vlines(6,0, 250, color='k',linestyle='--')
plt.legend()
ax.set_xlim(4,7)
ax.set_title('Count/Parallax for Locating Alpha Persei Cluster')

aP2 = aP1.query('parallax > 5.31 & parallax < 6')

fig, ax = plt.subplots()
aP3 = aP2.query('pmra < 32 & pmra > 18 & pmdec < -18 & pmdec > -32')
ax.scatter(data1['pmra'], data1['pmdec'], marker = '.', s=2, color = 'darkgrey') #alpha = 0.25)
ax.scatter(aP3['pmra'], aP3['pmdec'], marker = '.', s=2, color = 'darkmagenta', label = 'a Per Candidates')
ax.set_xlabel('PMRA [deg]')
ax.set_ylabel('PMDec [deg]')
ax.set_xlim(-100,100)
ax.set_ylim(-100,100)
ax.set_aspect(1)
circle = patch.Circle((23, -27), 8,linewidth=1.3, edgecolor='black', facecolor='none')
plt.gcf().gca().add_artist(circle)
ax.set_title('Proper Motion for Alpha Persei Cluster')
ax.legend(loc = (.01,.9),markerscale=10,prop={"size":15})

#aP3 = aP2.query('pmra < 32 & pmra > 20 & pmdec < -18 & pmdec > -32')
print(len(aP3))
```
